#### Vamos directamente a intentar crear el modelo para predecir

```python

from sklearn.model_selection import train_test_split

X = datos.data
y = datos.target 

X_ent, X_pru, y_ent_, y_pru = train_test_split(X, y, test_size=.3)
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score

modelo = svm.SVC() # Support VEctor Classifier
modelo.fit(X_ent, y_ent_)
predicciones = modelo.predict(X_pru)
print(f"Exactitud: {accuracy_score(y_pru, predicciones)}")
print(f"Presición: {precision_score(y_pru, predicciones)}")
```

```python
# Ahora pdemos empezar a ver cuales son los mejores hiperparametros que podemos ocupar en nuestros SVM

kernels = ["linear", "rbf", "sigmoid"]
gammas = [1.0, .01, .001, .0001, .00001]

print("Kernel:Gamma")
for kernel in kernels:
    for gamma in gammas:
        modelo = svm.SVC(kernel=kernel, gamma=gamma) # Support VEctor Classifier
        modelo.fit(X_ent, y_ent_)
        predicciones = modelo.predict(X_pru)
        print(f"{kernel}:{gamma:.5f} -> Exactitud: \t{accuracy_score(y_pru, predicciones)} \tPresición: \t{precision_score(y_pru, predicciones)}")
        # print(f"{kernel}:{gamma:.5f} -> Presición: \t{precision_score(y_pru, predicciones)}")
```

#### Ahora evaluamos desempeño

```python
# Podemos ver los resultados y vemos que el kernel lineal es de los mejores, y gamma da lo mismo
# Entonces:
modelo = svm.SVC(kernel="linear") # Support VEctor Classifier
modelo.fit(X_ent, y_ent_)
predicciones = modelo.predict(X_pru)
print(f"Exactitud: {accuracy_score(y_pru, predicciones)}")
print(f"Presición: {precision_score(y_pru, predicciones)}")
```

```python
# Se ve como mejoran los resultados, con esto listo escribiremos nuestro reporte de clasificaciones
from sklearn.metrics import classification_report
print(classification_report(y_pru, predicciones))
```

```python
# En general los resultados son bastante decentes (arriba)
# Podemos nuevamente hacer nuestra matriz de confusión
from sklearn.metrics import confusion_matrix
import pandas as pd

pd.DataFrame(confusion_matrix(y_pru, predicciones), columns=["Pred: No", "Pred: Si"], index=["Real: no", "Real: si"])
```

```python
# Se ven muy buenos resultados, incluso podemos dividir los conjuntos de entrenamiento
# y prueba la mitad y ver los resultados para ver como se coimporta SVM.SVC 
X_ent, X_pru, y_ent_, y_pru = train_test_split(X, y, test_size=.3)
modelo = svm.SVC(kernel="linear") # Support Vector Classifier
modelo.fit(X_ent, y_ent_)
predicciones = modelo.predict(X_pru)
print(f"Exactitud: {accuracy_score(y_pru, predicciones)}")
print(f"Presición: {precision_score(y_pru, predicciones)}")
print(classification_report(y_pru, predicciones))
print(pd.DataFrame(confusion_matrix(y_pru, predicciones), columns=["Pred: No", "Pred: Si"], index=["Real: no", "Real: si"]))
```
